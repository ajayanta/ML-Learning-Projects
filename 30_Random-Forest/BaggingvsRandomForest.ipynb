```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X,y=make_classification(n_features=5,n_redundant=0,n_informative=5,n_clusters_per_class=1)
```

```python
df=pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
```

```python
df['target']=y
```

```python
df.head()
```

```python
df.shape
```

```python
bc=BaggingClassifier(max_features=2)
```

```python
bc.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plot_tree(bc.estimators_[0])
```

```python
rclf =RandomForestClassifier(max_features=2)
```

```python
rclf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plot_tree(rclf.estimators_[0])
```

```python
# In bagging there is tree level feature sampling and in randomforest the column sampling is node level
```
